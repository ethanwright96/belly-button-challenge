```python
import json
import pandas as pd
```

```python
with open("data/samples.json") as file:
    data = json.load(file)
```

```python
df = pd.DataFrame(data)
df
```

```python
metadata_expanded = pd.json_normalize(df['metadata'])
metadata_expanded
```

```python
samples_expanded = pd.json_normalize(df['samples'])
samples_expanded
```

```python
df = pd.concat([df, metadata_expanded], axis=1)
df = pd.concat([df, samples_expanded], axis=1)
df
```
